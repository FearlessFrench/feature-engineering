```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
print(data.DESCR)
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
```

```python
df.head()
```

Diagnosis (M = malignant, B = benign)
รุนแรงกับไม่รุนแรง

```python
features = ['mean smoothness', 'mean compactness','mean concavity', 'mean concave points', 'mean symmetry']
```

```python
# สร้างคอลัมน์ใหม่ขึ้นมา โดยเอา features ใน list ข้างบนนี้มา sum กัน
df['added_features'] = df[features].sum(axis=1)
```

```python
df.head()
```

```python
# ถ้าจะเอาคอลัมน์ดังกล่าวมาคูณกัน
df['prod_features'] = df[features].prod(axis=1)
```

```python
df.head()
```

```python
# หา mean ก็ใช้ฟังก์ชัน mean
df['mean_features'] = df[features].mean(axis=1)
```

```python
df.head()
```

```python
# ปกติแล้วเขาจะหา sd คู่กับ mean ไปด้วย เพื่อความชัดเจนของข้อมูล
df['std_features'] = df[features].std(axis=1)
```

```python
df.head()
```

```python
# หรือจะหา max หา min ของคอลัมน์ดังกล่าว ก็ทำได้
df['max_features'] = df[features].max(axis=1)
df['min_features'] = df[features].min(axis=1)
```

```python
sns.violinplot(x="target", y="added_features", data=df)
plt.title('Added Features')
plt.show()
# ดูว่าระหว่าง 2 กลุ่มนี้ ข้อมูลมันสามารถแบ่งแยกกันได้มั้ย (เท่าที่ดูก็ยังแยกออกจากกันไม่ได้)
```

```python
sns.violinplot(x="target", y="prod_features", data=df)
plt.title('Product Features')
plt.show()
# ดูว่าระหว่าง 2 กลุ่มนี้ ข้อมูลมันสามารถแบ่งแยกกันได้มั้ย (เท่าที่ดูก็ยังแยกออกจากกันไม่ได้)
```

```python
sns.violinplot(x="target", y="mean_features", data=df)
plt.title('Mean Features')
plt.show()
# ดูว่าระหว่าง 2 กลุ่มนี้ ข้อมูลมันสามารถแบ่งแยกกันได้มั้ย (เท่าที่ดูก็ยังแยกออกจากกันไม่ได้)
```

```python
sns.violinplot(x="target", y="std_features", data=df)
plt.title('Std Features')
plt.show()
# ดูว่าระหว่าง 2 กลุ่มนี้ ข้อมูลมันสามารถแบ่งแยกกันได้มั้ย (เท่าที่ดูก็ยังแยกออกจากกันไม่ได้)
```
