```python
import pandas as pd
from feature_engine.imputation import MeanMedianImputer
from feature_engine.imputation import CategoricalImputer
from feature_engine.imputation import AddMissingIndicator
```

```python
data = pd.read_csv('creditApprovalUCI.csv')
```

```python
data.dtypes
```

```python
#หา Missing Data
data.isnull().mean()
```

```python
data.head()
```

```python
#เราจะแทนตรงที่เป็น Missing Data ด้วย 0 (ไม่เป็น Null), 1 (เป็น Null)
import numpy as np

for var in ['A1', 'A3', 'A4', 'A5', 'A6', 'A7', 'A8']:
    data[var + '_NA'] = np.where(data[var].isnull(), 1, 0)
    data[var + '_NA'] = np.where(data[var].isnull(), 1, 0)
```

```python
data.head()
```

```python
#ถ้าเราไม่ใช้ numpy สามารถใช้ Library นี้ได้
#imputer คือตัวโมเดล
imputer = AddMissingIndicator(variables=['A2', 'A11', 'A15'])
```

```python
imputer.fit(data)
```

```python
#apply โมเดลของเรากับ data ของเรา
data = imputer.transform(data)
```

```python
#เติมค่าที่สูญหายไปด้วย Median
#ถ้าอยากเติมด้วย mean() หรือค่าอื่นๆ ก็ทำได้เช่นกัน
for var in ['A2', 'A3', 'A8']:
    value = data[var].median()
    data[var] = data[var].fillna(value)
```

```python
data.head()
```

```python
#เติมค่าที่สูญหายแบบนี้ก็ได้ ใช้ฟังก์ชันของ Imputation ไม่ต้องวนลูปให้ยุ่งยาก
median_imputer = MeanMedianImputer(imputation_method='mean', variables=['A11', 'A15'])
```

```python
median_imputer.fit(data)
```

```python
data = median_imputer.transform(data)
```

```python
median_imputer.imputer_dict_
```

```python
#ถ้าตัวแปรเป็น Categorical Data ให้ใช้ฐานนิยม (Mode)
mode_imputer = CategoricalImputer(imputation_method='frequent', variables=['A4', 'A5', 'A6','A7'])
mode_imputer.fit(data)
data = mode_imputer.transform(data)
```

```python
mode_imputer.imputer_dict_
```

```python
#การเติมค่าที่สูญหายด้วยตัวเลขแบบสุ่ม (Arbitrary Number)
import pandas as pd
from feature_engine.imputation import ArbitraryNumberImputer
from feature_engine.imputation import CategoricalImputer
from feature_engine.imputation import RandomSampleImputer
```

```python
data = pd.read_csv('creditApprovalUCI.csv')
```

```python
data[['A2','A3', 'A8', 'A11']].max()
```

```python
#ทำให้เลขมันเป็น 99.999 หรือ เลขติดลบ เพื่อให้รู้ว่ามันเคยเป็น Missing Data มาก่อน
imputer = ArbitraryNumberImputer(arbitrary_number=9999999, variables=['A2','A3', 'A8', 'A11'])
imputer.fit(data)
data = imputer.transform(data)
```

```python
data.head()
```

```python
#แต่ถ้าไม่อยากให้มันเป็นเลขสุดโต่ง ก็ใส่อย่างอื่นลงไปได้
imputer = CategoricalImputer(variables=['A4', 'A5', 'A6','A7'], fill_value="missing")
```

```python
imputer.fit(data)
```

```python
data = imputer.transform(data)
```

```python
#เช็คดูว่าจำนวน Missing Data ลดลงไปมั้ย
data.isnull().mean()
```

```python
data[data["A4"] == "missing"]
```

```python
#วิธีการเติม Missing Data ด้วยการ Random ค่าใส่เข้าไป
imputer = RandomSampleImputer(variables=['A15'])
imputer.fit(data)
data = imputer.transform(data)
```
