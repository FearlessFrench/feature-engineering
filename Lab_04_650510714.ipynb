### 4. Construct new features (one or more) by combining two features with mathematical functions. Describe the meaning of the new feature in a comment.

```python
features = ['B.1.1.7 (Alpha)', 'B.1617.2 (Delta)', 'B.1.351 (Beta)', 'B.1.1.529 (Omicron)']
```

```python
# à¸—à¸³à¸à¸²à¸£à¸ªà¸£à¹‰à¸²à¸‡à¸„à¸­à¸¥à¸±à¸¡à¸™à¹Œà¹ƒà¸«à¸¡à¹ˆà¸—à¸µà¹ˆà¸„à¸³à¸™à¸§à¸“à¸•à¸²à¸¡à¸«à¸¥à¸±à¸à¸à¸²à¸£à¸„à¸“à¸´à¸•à¸¨à¸²à¸ªà¸•à¸£à¹Œ
data['added_features'] = data[features].sum(axis=1) # à¸„à¸³à¸™à¸§à¸“à¸œà¸¥à¸£à¸§à¸¡à¸à¸²à¸£à¹€à¸à¹‰à¸²à¸£à¸°à¸§à¸±à¸‡à¸‚à¸­à¸‡à¸—à¸±à¹‰à¸‡ 4 à¸ªà¸²à¸¢à¸žà¸±à¸™à¸˜à¹Œ
data['prod_features'] = data[features].prod(axis=1) # à¸„à¸³à¸™à¸§à¸“à¸œà¸¥à¸„à¸¹à¸“à¸à¸²à¸£à¹€à¸à¹‰à¸²à¸£à¸°à¸§à¸±à¸‡à¸‚à¸­à¸‡à¸—à¸±à¹‰à¸‡ 4 à¸ªà¸²à¸¢à¸žà¸±à¸™à¸˜à¹Œ

# à¸„à¸³à¸™à¸§à¸“à¸œà¸¥à¸•à¹ˆà¸²à¸‡à¸£à¸°à¸«à¸§à¹ˆà¸²à¸‡à¸ªà¸²à¸¢à¸žà¸±à¸™à¸˜à¹Œ Alpha à¸à¸±à¸šà¸ªà¸²à¸¢à¸žà¸±à¸™à¸˜à¹Œ Beta, Delta, Omicron à¸•à¸²à¸¡à¸¥à¸³à¸”à¸±à¸š
data['difference_Alpha_Beta'] = data['B.1.1.7 (Alpha)'].sub(data['B.1.351 (Beta)'])
data['difference_Alpha_Delta'] = data['B.1.1.7 (Alpha)'].sub(data['B.1617.2 (Delta)'])
data['difference_Alpha_Omicron'] = data['B.1.1.7 (Alpha)'].sub(data['B.1.1.529 (Omicron)'])
```
