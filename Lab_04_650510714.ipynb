### 3. Construct new features (one or more) by combining some features with statistical operations. Describe why you perform the statistical operation on the selected features in a comment.

```python
data.columns
```

```python
features = ['B.1.1.7 (Alpha)', 'B.1617.2 (Delta)', 'B.1.351 (Beta)', 'B.1.1.529 (Omicron)']
```

```python
# ทำการสร้างคอลัมน์ใหม่ที่คำนวณตามหลักการสถิติ
data['mean_features'] = data[features].mean(axis=1) # เพื่อหาว่าค่าเฉลี่ยของทั้ง 4 สายพันธ์รวมกันเป็นเท่าไร
data['std_features'] = data[features].std(axis=1) # เพื่อหาว่าทั้ง 4 สายพันธ์มีส่วนเบี่ยงเบนมาตรฐานเท่าไร
data['min_features'] = data[features].min(axis=1) # เพื่อหาว่าใน 4 สายพันธ์มีสายพันธ์ไหนมีการเฝ้าระวังน้อยที่สุด
data['max_features'] = data[features].max(axis=1) # เพื่อหาว่าใน 4 สายพันธ์มีสายพันธ์ไหนมีการเฝ้าระวังมากที่สุด
```

### 4. Construct new features (one or more) by combining two features with mathematical functions. Describe the meaning of the new feature in a comment.

```python
features = ['B.1.1.7 (Alpha)', 'B.1617.2 (Delta)', 'B.1.351 (Beta)', 'B.1.1.529 (Omicron)']
```

```python
# ทำการสร้างคอลัมน์ใหม่ที่คำนวณตามหลักการคณิตศาสตร์
data['added_features'] = data[features].sum(axis=1) # คำนวณผลรวมการเฝ้าระวังของทั้ง 4 สายพันธ์
data['prod_features'] = data[features].prod(axis=1) # คำนวณผลคูณการเฝ้าระวังของทั้ง 4 สายพันธ์

# คำนวณผลต่างระหว่างสายพันธ์ Alpha กับสายพันธ์ Beta, Delta, Omicron ตามลำดับ
data['difference_Alpha_Beta'] = data['B.1.1.7 (Alpha)'].sub(data['B.1.351 (Beta)'])
data['difference_Alpha_Delta'] = data['B.1.1.7 (Alpha)'].sub(data['B.1617.2 (Delta)'])
data['difference_Alpha_Omicron'] = data['B.1.1.7 (Alpha)'].sub(data['B.1.1.529 (Omicron)'])
```
